```python
x = df["OEMCode"].str.split(",")
x
```

## Create the required two columns
### Some rules for the required columns
**ISBN**
- if list have single value is directly set as ISBN
- elif 1st value starts with BBH the set 2nd value as ISBN
- else then set 1st value as ISBN 

**BBH Code**
- only those value starts with BBH in the list is set as BBH Code

```python
# BBH Code
# check for BBH code in 1st row
import re
bbh_code = list()
for k in range(x.shape[0]):
    for l in range(len(x[k])):
        bbh = x[k][l].strip()
    if re.findall("\ABBH", bbh):
        bbh_code.append(bbh)
    else:
        bbh_code.append("")
        
df["BBH Code"] = bbh_code
```

```python
# ISBN
# check the lenght of list equals to 1
isbn = list()
for n in x:
    if len(n) == 1:
        isbn.append(n[0])
    elif re.findall("\ABBH", n[0]):
        isbn.append(n[1])
    else:
        isbn.append(n[0])
        
df["ISBN"] = isbn
```
